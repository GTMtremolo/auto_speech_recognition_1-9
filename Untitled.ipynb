```python
from python_speech_features import mfcc, logfbank,delta
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM,Dense, Activation, Dropout
from keras.utils import to_categorical 
```

```python
data = pd.read_csv('data_speech_commands/test_data_1_9.txt', sep=" ", header=None)
raw_data = np.asarray(data)
target = []
for d in data[0]:
    target.append(d.split('/')[0])
y_test = np.asarray(target)

x_test = np.zeros((len(raw_data),99,26))
for i in range(len(raw_data)):
    (rate,sig) = wav.read("data_speech_commands/"+raw_data[i][0])
    mfcc_feat = mfcc(sig,rate, numcep=26 )
    mfcc_feat = delta(mfcc_feat, 2)
    
    while True:
        if len(mfcc_feat) == 99:
            break
        mfcc_feat = np.append(mfcc_feat, np.zeros((1,26)),axis=0)
    x_test[i] = mfcc_feat
print('done')
```

```python
print(x_test.shape)
for i in range(len(y_test)):
    if y_test[i] == 'one':
        y_test[i] = 1
    if y_test[i] == 'two':
        y_test[i] = 2
    if y_test[i] == 'three':
        y_test[i] = 3
    if y_test[i] == 'four':
        y_test[i] = 4
    if y_test[i] == 'five':
        y_test[i] = 5
    if y_test[i] == 'six':
        y_test[i] = 6
    if y_test[i] == 'seven':
        y_test[i] = 7
    if y_test[i] == 'eight':
        y_test[i] = 8
    if y_test[i] == 'nine':
        y_test[i] = 9
y_test= y_test.astype(np.uint8)
```

```python
data = pd.read_csv('data_speech_commands/train_data_1_9.txt', sep=" ", header=None)
raw_data = np.asarray(data)
target = []
for d in data[0]:
    target.append(d.split('/')[0])
    
y_train = np.asarray(target)

x_train = np.zeros((len(raw_data),99,26))
for i in range(len(raw_data)):
    (rate,sig) = wav.read("data_speech_commands/"+raw_data[i][0])
    mfcc_feat = mfcc(sig,rate,numcep= 26)
    mfcc_feat = delta(mfcc_feat, 2)
    while True:
        if len(mfcc_feat) == 99:
            break
        mfcc_feat = np.append(mfcc_feat, np.zeros((1,26)),axis=0)
    x_train[i] = mfcc_feat
print('done')
```

```python
for i in range(len(y_train)):
    if y_train[i] == 'one':
        y_train[i] = 1
    if y_train[i] == 'two':
        y_train[i] = 2
    if y_train[i] == 'three':
        y_train[i] = 3
    if y_train[i] == 'four':
        y_train[i] = 4
    if y_train[i] == 'five':
        y_train[i] = 5
    if y_train[i] == 'six':
        y_train[i] = 6
    if y_train[i] == 'seven':
        y_train[i] = 7
    if y_train[i] == 'eight':
        y_train[i] = 8
    if y_train[i] == 'nine':
        y_train[i] = 9
y_train= y_train.astype(np.uint8)
```

```python
y_train = to_categorical(y_train)
Y_test = to_categorical(y_test)
print(x_test.shape, y_test.shape, x_train.shape, y_train.shape)
```

```python
print(y_test[500])
```

```python
model = Sequential()
model.add(LSTM(units=26,input_shape=(None,26), activation='tanh', recurrent_activation='sigmoid'))
model.add(Dropout(0.30))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.30))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train,y_train,batch_size=128 ,epochs=700,verbose=1,validation_data=(x_test,Y_test))
model.save("asr.h5")
```

```python
plt.plot(history.history['loss'])
plt.show()
```

```python
result = model.predict(x_test)
print(result.shape)
for i in range(len(result)):
    print(np.argmax(result[i]))
```
